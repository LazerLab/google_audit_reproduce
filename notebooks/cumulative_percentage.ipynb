# Cumulative top domain counts

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
```

```python
domain_counts = pd.read_csv('../data/house_analysis/house_domain_counts_unique.csv')
```

```python
domain_counts  = domain_counts.sort_values(by=['counts'], ascending=False)
```

```python
domain_counts["cumulative_counts"] = domain_counts['counts'].cumsum()
```

```python
sum_domain = domain_counts['counts'].sum()
```

```python
domain_counts["cum_proportion"] = domain_counts['cumulative_counts'] / sum_domain
```

```python
domain_counts["top"] = range(0,len(domain_counts))
```

```python
fig = plt.figure(figsize = (12, 8))
plot_df = domain_counts

plt.plot(plot_df.top, plot_df['cum_proportion']*100)
plt.xticks(rotation=50, fontsize=20)
plt.yticks(fontsize=20)
 
plt.xlabel("Top n domains",fontsize=25)
plt.ylabel("% of results",fontsize=25)
plt.title("Cumulative percentage of domains",fontsize=25)


plt.savefig("../data/plots/house_cum_top_n_domains.png", bbox_inches='tight', dpi=200) 
plt.show()
```
